```python
#Curve fit
x = np.linspace(0,4,100)
y = 1/3*x**3-3/5*x**2+2+np.random.randn(x.shape[0]) # on va creer une fonction cubique avec de noise
plt.scatter(x,y)
def f(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d
from scipy import optimize
params , params_cov=optimize.curve_fit(f,x,y) # determiner les coefficients (params) les plus proches de y généré en random , params_cov: covariance
plt.scatter(x,f(x,params[0],params[1],params[2],params[3]))
```

```python
# minimize
def f(x):
    return x**2+15*np.sin(x)
x = np.linspace(-10,10,100)
x0 = 0
result = optimize.minimize(f,x0=x0).x # trouver un minimum locale de la fonction f le plus proche du point x0
print(result)
plt.plot(x,f(x),lw=3,zorder=-1)
plt.scatter(result,f(result),s=100,c='r',zorder=1)
plt.scatter(x0,f(x0),s=200,marker='+',c='g',zorder=1)
```

```python
# TRAIN TEST SPLIT
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
X = iris.data
# X: 
# _________________________________________
# | lgr sep | lrg sep | lgr pet | lrg pet |
# -----------------------------------------
# |  ...    |   ...   |  ...    |  ...    | 
# -----------------------------------------
X
y = iris.target.reshape(iris.target.shape[0],1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# cross validation
# consiste a entrainer puis valider
# notre modele sur plusieurs decoupes
# possible de trainset
from sklearn.model_selection import cross_val_score
#ici on veut comparer deux modeles avec des parametres diff
model1 = KNeighborsClassifier(n_neighbors=1)
model2 = KNeighborsClassifier(n_neighbors=2)
#cross_val_score[i] = score% de modele en division i
# on peut par suite calculer la moyenne dans les 5 division (cv=5)
# et comparer le resultat
print(cross_val_score(model1,X_train,y_train,cv=5,scoring='accuracy'))
print(cross_val_score(model2,X_train,y_train,cv=5,scoring='accuracy'))
# cv: nombres de splits que l'on veut
# scoring: metrics ('accuracy' generalement) 
```

```python
from sklearn.model_selection import validation_curve
model = KNeighborsClassifier()
domaine = np.arange(1,50) # le domaine des valeurs possibles de parametres qu'on souhaite optimiser
# on veut choisir le meilleur n_neighbors parametre
# pour ce faire: on utilise la methode validation_curve()
# cette methode va retourner deux tableaux:
# val_score: contient les resultats
# de modele selon les 5 divisions de validation (cv = 5)
# val_score.shape = (k,cv) c'est a dire: pour chaque nombre i tq n_neighbors = k
# on a un modele(n_neighbors = k) et son resultat selon les (cv=5) divisions
# train_score: de meme, sauf que le modele est entrainé sur les X_train et y_train
train_score,val_score=validation_curve(model,X_train,y_train,param_name='n_neighbors',param_range=domaine,cv=5)
plt.plot(domaine,val_score.mean(axis=1),label='validation')
plt.plot(domaine,train_score.mean(axis=1),label='train')
plt.ylabel('score')
plt.xlabel('n_neighbors')
plt.legend()
```

```python
# dans ce qui est passé,
# on a chercher a optimiser un seul parametre
# (n_neighbors de KNeighborsClassifier),
# MAIS il existe d'autre parametres dans ce modele
# on souhaite a chercher la combinaison optimale
# des parametres pour ce modele ??
# => GridSearchCV
# brute force search des parametres
from sklearn.model_selection import GridSearchCV
param_grid={ # contient les parametres qu'on souhaite optimiser ainsi que le domaine de leurs valeurs
    'n_neighbors': np.arange(1,20),
    'metric':['euclidean','manhattan']
} # 'n_neighbors' et 'metric' sont des parametres relatives au KNeighborsClassifier
# si on a un autre modele, on aura different parametres à optimiser
grid = GridSearchCV(KNeighborsClassifier(),param_grid,cv=5)
grid.fit(X_train,y_train)
# grid.best_params_ : parametres optimales
# grid.best_score_ : top score de modele
model = grid.best_estimator_ # modele avec les parametres optimales
model.score(X_test,y_test)
```

```python
# confusion matrix:
# comparer les resultats reelles 
# et les resultats de modeles 
# c'est une matrice carre de taille NxN
# tq : N c'est le nombres de classes
# de notre target
# ici on a 3 classes donc 3x3 ['setosa', 'versicolor', 'virginica']
# chaque ligne de la matrice represente les classes des echantillion predictes par le modele
# par exemple:
# 9 0 0 -> 9 valeurs de la classe 1 sont mis correctement dans la classe 1
# 0 7 3 -> 7 valeurs de la classe 2 sont mis correctement dans la classe 2 et 3 valeurs sont mis faussement dans la classe 3
# 1 0 10 -> 10 valeurs de la classe 3 sont mis correctement dans la classe 3 et 1 valeur est mis faussement dans la classe 1
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model.predict(X_test)) 
```

```python
# Learning curve
# est ce que le modele pourrait il s'ameliorer
# en lui donnant plus de données?
# pour repondre: en doit tracer le learning curve
# sachant qu'on est sur que le modele doit atteindre
# un plaffond maximale de learning et on n'a plus
# besoin de donnees pour l'ameliorer
from sklearn.model_selection import learning_curve
N,train_score,val_score=learning_curve(model,X_train,y_train,train_sizes=np.linspace(0.1,1.0,10),cv=5)
# train_sizes: un tableau contenant des pourcentages de quantite
# de donnees qu'on souhaite entrainer le modele a chaque fois
# exemple: train_sizes=[20%,30%,40%...]
print(train_score)
plt.plot(N,train_score.mean(axis=1),label='train')
plt.plot(N,val_score.mean(axis=1),label='validation')
plt.xlabel('train_sizes')
plt.legend()
```

```python
# <<<<<<<<<<<<<<SOLUTION>>>>>>>>>>>>>>
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
titanic = sns.load_dataset("titanic")
# titanic.drop(['alone','alive','who','adult_male','embark_town','class','sibsp','parch','fare','embarked','deck'],axis=1,inplace=True)
titanic = titanic[['survived','pclass','sex','age']]
titanic.dropna(axis=0,inplace=True)
titanic['sex'].replace(['male','female'],[0,1],inplace=True)
X = titanic[['pclass','sex','age']]
y = titanic['survived']
# 1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train,y_train)
from sklearn.metrics import confusion_matrix
print(" 1)")
print("Confusion Matrix:\n",confusion_matrix(y_test,model.predict(X_test)) )
print("\nModel score after training:")
print(model.score(X_test,y_test))
# 2)
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_neighbors':np.arange(1,20),
    'metric':['euclidean','manhattan'],
    'weights':['distance', 'uniform'],
    'algorithm':['auto','ball_tree','kd_tree','brute']
}
grid = GridSearchCV(KNeighborsClassifier(),param_grid,cv=30,scoring='accuracy')
grid.fit(X_train,y_train)
print(" 2)")
print("Grid search resultats:")
print("Top score:\n",grid.best_score_)
print("Avec parametres:\n",grid.best_params_)
model = grid.best_estimator_
# 3) Oui, en effet:
print(" 3)")
from sklearn.model_selection import learning_curve
N,train_score,val_score=learning_curve(model,X_train,y_train,train_sizes=np.linspace(0.1,1.0,30),cv=5)
plt.plot(N,train_score.mean(axis=1),label='train')
plt.plot(N,val_score.mean(axis=1),label='validation')
plt.xlabel('train_sizes')
plt.legend()
```
